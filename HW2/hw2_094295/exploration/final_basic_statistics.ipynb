```python
import os
from glob import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

First we will try to understand how many examples of each class we have. In order to understand the starting balance of the data.

```python
total = 0
easy_view_dict = dict()
for folder in sorted(glob('/home/student/Data-analysis-and-presentation/HW2/hw2_094295/final_data/train/*')):
    numeral = folder.split(sep='/')[-1]
    # print(numeral)
    folder_size = len(glob(folder+'/*.png'))
    # print(folder_size)
    total += folder_size
    easy_view_dict[numeral] = folder_size
print(f"Total of {total} pictures")
pd.DataFrame.from_dict(data=easy_view_dict, orient='index')
```

As we can see, we do not exceed the 10,000 limit at first. Therefore we will use this to create a baseline.
